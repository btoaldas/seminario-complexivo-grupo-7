# Expresiones

## ¿Qué son las expresiones en Python?

Son el mecanismo para hacer cálculos.

Básicamente las expresiones se componen de combinaciones de **valores** e **identificadores** con operadores.

Las **expresiones** combinan:
* Variables
* Valores
* Operadores
* Paréntesis
* Nombres de funciones especiales (raíz cuadrada, logaritmo, etc.)

Toda expresión tiene un valor que es el resultado de evaluarla de **izquierda a derecha**, tomando en cuenta las precedencias, es decir, lo que sucedió anteriormente.

#### Ejemplo 7
Imaginémonos si tenemos ya nombres de variables como `items` y necesitamos realizar varios cálculos.

Esta sería la forma habitual.

```python
# La variable "items" almacena el número de algunos elementos
items = 10

items = items + 5 # sumamos 5 a 10 (el valor anterior de item)
print(items) # nuevo valor

items = items - 5 # restamos 5 de 15
print(items) # nuevo valor

items = items * 3 # multiplicamos por 3
print(items) # nuevo valor

items = items / 5 # dividimos entre 5
print(items) # nuevo valor
```

Ahora esta sería la forma abreviada:

```python
# La variable "items" almacena el número de algunos elementos
items = 10

items += 5 # igual que items = items + 5
print(items)

items -= 5 # igual que items = items - 5
print(items)

items *= 3 # igual que items = items * 3
print(items)

items /= 5 # igual que items = items / 5
print(items)
```

Utilizamos menos símbolos, y obtenemos los mismos resultados.

### Orden de las operaciones (precedencia)

**IMPORTANTE**

---
Para resolver expresiones con múltiples operadores, incluyendo diferente tipo de operador, aritmético, booleano o relacional, se debe seguir un **orden de prioridad** para realizar primero una operación y luego otra.


En caso de que haya dudas de realizar una operación, es recomendable utilizar **paréntesis**.

El órden de prioridad o precedencias depende del tipo de lenguaje de programación.
* La máxima prioridad la tiene la realización de las expresiones entre **paréntesis**.
* Luego la **exponenciación** y las operaciones **unarias**.
* Después siguen las operaciones **aritméticas**, donde se priorizan la multiplicación y división sobre la suma y resta:
  * PEMDAS (Paréntesis, Exponenciación, Multiplicación - División, Adición - Sustracción).
* Después siguen las operaciones **relacionales** y por último las **booleanas**, pero con mayor prioridad de la negación **`not`**, luego **`and`** y por último **`or`**

# Errores

En cualquier situación se pueden dar los errores a nivel de lenguaje de programación, sea porque como programadores cometimos un error de **tipeo** o porque pedimos algo **imposible** a Python.

Cualquiera de las dos situaciones, hay que saber identificar los distintos **errores**.

## Ejemplo 17
¿Qué va a pasar en la siguiente ejecución? ❓

```python
777/0
```

En este mensaje de error, vemos la siguiente explicación `ZeroDivisionError`, el cual nos indica que no se puede dividir entre cero.

Tomar en cuenta que **todos** podemos cometer **errores** al momento de programar, por lo que no nos debemos sentir mal, o frustrarnos, sino es una forma de indicarnos el programa que algo salió mal, y podemos corregirlo inmediatamente.

## ¿Cómo manejar los errores?

Cuando Python encuentra un error, termina el programa.

Esto también pasa cuando estás jugando a un videojuego o utilizando una aplicación, y el programa de repente se bloquea.

Podemos **evitar** errores utilizando el operador para procesar excepciones **`try-except`**:

### Ejemplo 18
Utilicemos en el ejemplo anterior `try-except`

```python
try:
  print(777 / 0)
except:
  print('¡No podemos dividir entre cero!')
  print('¡Sigamos!')
```

```python
try:
  print(7/1)
except:
  print("no se puede dividir para cero")
  print("programando")
```

```python
a = 77
c = "f"
```

```python
try:
    z = a / c
except ZeroDivisionError:
    z = 0
except SyntaxError:
  print("está mal la sintaxis")
except:
  print("revisa las variables")
```

```python
print(z)
```

Básicamente, Python intentará ejecutar el código desde el bloque `try`, si hay un error, ejecutará el código del bloque `except`.

**Regla de sintaxis**

---
Tomar en cuenta que tanto `try` como `except` van seguidos de dos puntos. Los bloques con instrucciones se escriben después de cada uno.

Cada línea dentro de los bloques está con sangría.

Python utiliza dos puntos y sangrías para entender dónde termina un bloque y comienza otro.

Un espacio extra o faltante **provocará un error**.
