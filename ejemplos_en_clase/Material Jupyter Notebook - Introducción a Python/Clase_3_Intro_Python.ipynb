# Strings

La estructura de datos `str` (string) es una cadena de caracteres de cualquier tipo:
* Letras
* Números
* Caracteres especiales ($, &, /, etc.)

Básicamente cualquier caracter Unicode (Universal Character Encoding Standard
utilizado para la representación de texto para el procesamiento informático).

Para crearlo se debe encerrar entre comillas simples o dobles (' ', " ").

Una vez creado un `str`, es una constante, resultando ser un objeto **inmutable** (**estático** o de **valores/tamaño fijos**).
* Un objeto inmutable se caracteriza porque los elementos **no pueden ser cambiados, ni eliminados, ni añadir nuevos elementos.**

Para entender esto de mejor manera, hay que explicar primero qué son los índices de los `str`.

## Índices

En Python cada caracter de una secuencia tiene un **índice**, es decir, un valor que representa la **posición** del caracter.

* Toda secuencia de indexación comienza con **cero** (0)
  * La primera posición no se representa con el uno, sino con el cero, y así de forma consecutiva.

### Indexación negativa

Tomar en cuenta que Python también brinda una forma práctica de indexar comenzando desde el final de un `str`, el cual se realiza con **números negativos**.
* El último elemento de un `str` inicia con -1, el penúltimo con -2, y así de forma consecutiva.

A continuación vemos un ejemplo visual de la palabra `everyone can` junto con sus índices, positivos y negativos.

## Slices

**Los slices permiten acceder a varios caracteres al mismo tiempo**.

Además de acceder a un caracter individual, es importante saber cómo acceder a varios al mismo tiempo, esto se llama **slicing**.

Para cortar un string, se coloca el número del índice inicial y el número del índice final dentro de los corchetes, separándolos con dos puntos.

### Ejemplo 10
¿Qué resultado va a salir en el siguiente slicing? ❓

```python
ciudad = 'Moscú, Rusia'
substraer = ciudad[3:8]
print(substraer)
```

```python
print(ciudad[3:])
```

```python
print(ciudad[:3])
```

```python
print(ciudad[:10])
```

```python
len(ciudad)
```

**IMPORTANTE**

---
El último caracter del rango no forma parte del resultado final.

### Ejemplo 11

¿Qué resultados se obtendrá en los siguientes slices? ❓

```python
ciudad = 'Moscú, Rusia'
print(ciudad[3:1000])
```

```python
print(ciudad[-15:500])
```

```python
print(ciudad[3:0])
```

Python ignora los valores del índice que están fuera del rango real:
* El slice `ciudad[3:1000]` contendrá los caracteres desde el índice `3` hasta el final del string.
* `ciudad[-15:500]` contendrá el string completo, ya que el rango incluye todos los índices.
* `ciudad[3:0]` como el primer índice es mayor que el último, devolverá un string vacío.

Para cortar un slice a partir del primer caracter, no es necesario indicar su índice, se lo puede realizar de la siguiente forma:

```python
ciudad[:4]
```

De igual forma se puede omitir el índice del último caracter si se quiere que el slice incluya todo hasta el final del string.

```python
ciudad[7:]
```

## Métodos de string

### ¿Qué es un método?

Python tiene un conjunto de **métodos** integrados para procesar strings.

Los métodos son como las funciones, pero solo se pueden llamar a los métodos dentro de un objeto determinado, y mediante la notación de puntos.

La **función** tiene la siguiente notación: `funcion(argumento)`.

El **método** se utiliza de la siguiente forma: `objeto.metodo(argumento)`.

**IMPORTANTE**

---
Un método **afecta** el contenido del objeto mientras que una función te **entrega información** sobre el objeto.

### ¿Cómo ayudan los métodos?

El conocimiento de los métodos integrados puede ayudarte a evitar la necesidad de crear funciones similares por tu cuenta.

Aquí están algunos métodos útiles:
* `replace()` - reemplaza caracteres o secuencias de caracteres en el string.
* `find()` - busca un substring en un string.
* `upper()` - hace que todas las letras de un string se conviertan a mayúsculas.
* `lower()` - hace que todas las letras de un string se conviertan a minúsculas.

### `replace()`

El método `replace()` toma dos argumentos:
1. El substring que se reemplazará.
2. El substring de reemplazo.

#### Ejemplo 18
Veamos el siguiente ejemplo con `replace()`.

```python
# La variable "some_string" almacena algo de texto.
some_string = 'leaving'
print(some_string)
```

```python
some_string = some_string.replace('leav', 'stay') # reemplazando la substring 'leav' con 'stay'
print(some_string)
```

```python
word = "casa"
print(word)
word = word.replace("a", "i", 2)
print(word)
```

```python
word = "casa"
print(word)
word = word.replace("a", "e", 1)
print(word)
```

El método devolverá un nuevo objeto de string modificado.

Puede almacenarse en la variable original, en una variable nueva o simplemente mostrar en la pantalla.

### `find()`

El método `find()` busca un substring dentro de un string. Si lo encuentra, devuelve el índice en el que comienza.

Solo se pasa un argumento al método, el substring que estás buscando.

#### Ejemplo 19
Veamos el siguiente ejemplo con `find()`.

¿Qué resultado creen que aparezca? ❓

```python
frase= 'Leer, leer, leer, vivir la vida'
idx = frase.find('leer')
print(idx)
```

El substring leer aparece varias veces en la string, pero el método devolverá solo el índice de su primera instancia.

También se puede establecer un rango de índices para `find()`.

#### Ejemplo 20
¿Cuál será el índice que aparecerá? ❓

```python
frase= 'Leer, leer, leer, vivir la vida'
idx = frase.find('leer', 10, 20)
print(idx)
```

#### Ejemplo 21
¿Qué devolverá? ❓

```python
french_greeting = 'Bonjour'
idx = french_greeting.find('soir')
print(idx)
```

Si `find()` no encuentra nada, devolverá `-1`.

### `upper()` y `lower()`

```python
word = 'AbRacaDABRA'
lower_word = word.lower()
upper_word = word.upper()
print(lower_word, upper_word)
```

## F-string

### Usando variables dentro de un string

Tenemos el siguiente ejemplo donde tenemos tres variables `nombre`, `edad`, `altura`.

```python
nombre = "Max"
edad = 23
altura = 175.7
```

Para poder escribir lo siguiente: `Este es 'nombre'. Tiene 'edad' años y mide 'altura' cm.` debemos transformar las variables numéricas a `str`.  

```python
introduccion = "Este es " + nombre + '. Tiene ' + str(edad) + ' años y mide ' + str(altura) + ' cm.'
print(introduccion)
```

```python
introduccion = "Este es " + nombre + '. Tiene ' + edad + ' años y mide ' + altura + ' cm.'
print(introduccion)
```

Funciona, pero este código es compliucado de escribir y de leer porque:
* El operador `+` se debe utilizar varias veces para unificar todo.
* Uso de la función `str()` para que realmente funcione la unión.
* Se debe poner espacios extras al inicio y final para que la unión sea correcta.

### Uso correcto de F-string

Los F-strings está precedido por la letra `f` (**formateado**).

Los nombres de las variables se pueden indicar entre llaves `{}`, en vez de unir con `+`.

No hay que preocuparse por convertir los números en strings, ya que lo hará **automáticamente**.

```python
nombre = "Max"
edad = 23
altura = 175.7
```

```python
introduccion = "Este es " + nombre + '. Tiene ' + str(edad) + ' años y mide ' + str(altura) + ' cm.'
print(introduccion)
```

```python
introduccion = f"Este es {nombre}. Él tiene {edad} años y mide {altura} pies."
print(introduccion)
```

```python
introduccion = f"Este es {nombre.upper()}. Él tiene {edad + 1} años y mide {altura} cm."
print(introduccion)
```

### Operaciones dentro de un F-string

Las operaciones como los métodos de string o los cálculos aritméticos también se pueden llamar dentro de las llaves:

#### Ejemplo 22
Veamos un ejemplo con F-string y métodos.

```python
nombre = 'Max'
edad = 23

introduccion = f"Este es {nombre.upper()}, y el año que viene tendrá {edad + 1}."
print(introduccion)
```

### Formateo de números

En el caso de las F-strings, puedes cambiar la forma en la que se muestran los números.

Para formatear un porcentaje, hay que utilizar la expresión `.n%` después de la variable, donde `n` es el número de lugares decimales.

#### Ejemplo 23
Veamos cuánto es el porcentaje de la cuota de internet de los rusos con uno y dos decimales.

```python
ru_website_share = 0.0612
print(f'La cuota de los sitios rusos en internet es {ru_website_share:.1%}.')
print(f'La cuota de los sitios rusos en internet es {ru_website_share:.2%}.')
print(f'La cuota de los sitios rusos en internet es {ru_website_share:.5%}.')
```

### Sin F-strings

Este estilo de escritura de strings es bastante nuevo en Python. Apareció en la versión 3.6.

Anteriormente, las strings se formateaban con el método integrado `format()`.

Hoy día, el método `format()` se considera anticuado. Sin embargo, es útil comprender esta sintaxis si se encuentran con un código más antiguo.

La estructura es casi la misma, solo que en lugar de colocar la variable, hay que colocar la posición dentro del método `format()`.

Observen que ya no escribimos `f` al principio del string y que la numeración de la posición comienza con `0`, al igual que con la indexación.

#### Ejemplo 24
Ejemplo sin F-Strings

```python
nombre = 'Max'
edad = 23
altura = 175.2

introduccion = 'Este es {0}. Él tiene {1} años y mide {2} cm.'.format(nombre, edad, altura)
print(introduccion)
```

```python
nombre = 'Max'
edad = 23
altura = 175.2

introduccion = 'Este es {0}. Él tiene {1} años y mide {2} cm.'.format(nombre.upper(), edad+1, altura)
print(introduccion)
```
