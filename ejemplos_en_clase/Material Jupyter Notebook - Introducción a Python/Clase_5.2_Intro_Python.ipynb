# Condicionales `if`, `elif` y `else`

Hay que tomar en cuenta que hasta ahora se han creado `algoritmos lineales` en los que las líneas de código simplemente **se ejecutan una tras otra**.

Sin embargo, no todos los algoritmos funcionan de esta manera. A veces, Python sigue ciertas instrucciones mientras ignoran otras.
  * En otras palabras, el código puede contener **ramas**.
  * Se puede crear estas ramas dentro del código utilizando los operadores condicionales `if` y `else`.

A continuación se puede observar un ejemplo visual, de la siguiente pregunta:

## Sintaxis

Una expresión condicional en Python contiene un encabezado y un cuerpo.
  * El encabezado contiene la condición (una expresión lógica).
  * El cuerpo contiene instrucciones (ramas de código) con una sangría de cuatro espacios.

esta_lloviendo.png

Si la expresión lógica devuelve `True`, el operador `if` ejecutará los comandos desde el cuerpo de la condición.

Tomar en cuenta que en caso de continuar evaluando distintas opciones, se escribe `elif` que se refiere a else if, donde en Python se escribe de forma abreviada.

### Ejemplo 11
Cambiemos el estado del clima para ver cómo afecta el condicional.

```python
# se ingresa información por parte del usuario
clima = input("¿Cuál es el estado del clima? ").lower()
# resultados validos
condiciones_validas = ["lluvia", "lloviendo", "llueve", "lluvioso"]

# condición
if clima in condiciones_validas:
    print("Toma un paraguas.")
else:
    print("No tomes el paraguas.")
```

### Ejemplo 12
Creemos un programa que convierte °C a °F

```python
# ingresar una temperatura en celcius
C = float(input("Ingrese la temperatura en ºC: "))

# condición 
if C < -273.15:
  print('¡Temperatura en ºC irreal por debajo del cero absoluto!')
else:
  # convierte la temperatura a Fahrenheit
  F = 1.8*C + 32
  print(f"{C} ºC equivale a {F:.2f} ºF")
```

### Ejemplo 13
Creemos un programa que convierte una nota numérica a cualitativa para la clase "Analítica con Python".

```python
print('Programa que convierte una nota numérica a cualitativa')
nota = float(input('Nota numérica del 0 al 100 (puede ser con 2 decimales): '))

if 0 <= nota < 60.99:
  calif = 'No aprueba el curso (F)'
elif 61 <= nota < 70.99:
  calif = 'Deficiente (D)'
elif 71 <= nota < 80.99:
  calif = "Regular (C)"
elif 81 <= nota < 90.99:
  calif = 'Bueno (B)'
elif 91 <= nota < 100:
  calif = 'Sobresaliente (A)'
else:
  calif = "No está dentro del rango de 0 a 100."

print(f'La nota {nota} equivale a {calif}')
```

# Bucle `While`

## ¿Qué es el bucle `while` o `while` loop?

Al igual que el `for` loop, el `while` loop es usado para repetir un bloque específico de código, que suele ser una operación o instrucción.

Sin embargo, en la estructura del bucle `while`, el bloque de secuencia de instrucciones se **repite** siempre que la condición dada por el valor de una variable o expresión booleana **sea cierta (`True`)**.

## Sintaxis del bucle `while`

    while verificacion_de_expresion:
        cuerpo del while

Tenemos lo siguiente:
* La palabra clave `while`.
* La verificación de la expresión que es una operación booleana.
  * Solo entra al bucle o loop si el resultado es `True`.
  * En caso de que el resultado de la operación booleana sea `False`, deja de iterar o entrar al bucle.
* Los dos puntos `:` es un elemento esencial que separa las dos partes del bucle.
* El cuerpo del `while`.
  * Constituye el cuerpo del bucle.
  * El cuerpo contiene comandos que el bucle ejecuta en cada iteración del ciclo.
  * Cada comando en el cuerpo del bucle está precedido por una sangría de cuatro espacios, lo que indica que el comando debe ejecutarse dentro del bucle separado del resto del código.

### Ejemplo 14
Realizar un código con `while` que indique los `N` primeros números naturales en forma descendicente.

```python
# usuario ingresa número
N = int(input("Entra un número natural: "))
# i igual a número
i = N

# while loop, se detiene solo si es igual a 1
while i >= 1:
  print(i, end=" ")
  i -= 1    # se decrece menos 1, i = i - 1
print(f"\nHemos mostrado {N} números naturales decreciendo")
```

### Ejemplo 15
Realiza un código con `while` loop que al ingresar un número entero, calcule la suma de los `n` primeros números naturales.

```python
# número ingresado
n = int(input('Entra un número natural: '))
# i es igual a 1
i = 1
# suma inicia en 0; contador
suma = 0

# while loop que inicia en 1 y se detiene si es igual a n
while i <= n:
  suma = suma + i # suma += i
  i = i + 1

# imprime el resultado final
print(f'La suma de los números naturales hasta {n} es {suma}')
```

Si queremos entender qué proceso es el que realiza dentro del bucle, podemos hacer lo siguiente.

```python
# número ingresado
n = int(input('Entra un número natural: '))
# i es igual a 1
i = 1
# suma inicia en 0; contador
suma = 0

# while loop que inicia en 1 y se detiene si es igual a n
while i <= n:
  print(f"Iniciamos en {suma} la sumatoria, y le añadimos {i}.")
  suma += i # suma = suma + i
  i += 1    # i = i + 1
  print(f"El resultado final de la iteración {i-1} es: {suma}.\n")

# se imprime el mensaje
print(f'La suma de los números naturales hasta {n} es {suma}')
```

Generalmente se suele utilizar más el bucle `for`, sin embargo, es importante saber el bucle `while` en caso de que se quiera repetir una tarea hasta que el resultado sea `False`.

### Ejemplo 16
Crea un programa en Python que calcule el saldo de tu **cuenta de ahorros** después de restar tus **gastos del mes de septiembre**. El programa debe pedir al usuario la cantidad en la cuenta de ahorros y los gastos del mes, y luego mostrar el **saldo restante**.

Apliquemos todo lo que hemos aprendido hasta ahora:

```python
# while loop que solo se detiene si es False
while True:
    try:
        # usuario ingresa cantidad en cuenta de ahorros
        cuenta_ahorros = float(input("¿Cuánto tienes en tu cuenta de ahorros? "))
        # usuario ingresa gastos del mes
        gasto_mes = float(input("¿Cuánto has gastado en el mes de septiembre? "))
        break  # si los datos son válidos, sale del bucle
    except ValueError:
        # Se ejecuta si se ingresa un valor no entero, mostrando un mensaje de error
        print("Por favor ingresa un número entero válido.")

# Verifica si los gastos del mes son cero
if gasto_mes == 0:
    # Si no hay gastos, muestra el saldo actual de la cuenta
    print(f"Fondos cuenta de ahorro: ${cuenta_ahorros:,.2f}")
else:
    # Resta los gastos del saldo de la cuenta
    cuenta_ahorros -= gasto_mes
    # Verifica si la cuenta está en negativo
    if cuenta_ahorros < 0:
        # Si es negativo, muestra cuánto debe el usuario
        print(f"Debes: ${cuenta_ahorros:,.2f}")
    else:
        # Si es positivo, muestra el saldo restante
        print(f"Fondos cuenta de ahorro: ${cuenta_ahorros:,.2f}")
```
