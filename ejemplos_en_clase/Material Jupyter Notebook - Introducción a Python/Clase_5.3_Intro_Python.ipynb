# Funciones

En programación, una **función** funciona como una receta.

La idea básica escribir un conjunto de instrucciones y guardarlas con un nombre.
* Cuando se llama a la función, Python ejecuta las instrucciones y devuelve el resultado.

Ventajas del uso de funciones:
* Es más fácil de leer el código.
* Se eliminan las repeticiones en el código.
* Se puede utilizar cuantas veces queramos.

## Sintaxis

Los cuatro componentes principales de una función son:
* El nombre de la función.
* Los argumentos que cambian el comportamiento de la función.
* Las instrucciones que se ejecutan cuando se llama a la función (también llamadas el cuerpo de la función).
* El resultado que devuelve la función.

Aquí está un ejemplo de cómo se debe escribir una función:

funciones.png

```python
def omelet(eggs_number = 2, eggs_broken = 0):
    result = f"The omelet is ready! Eggs used: {eggs_number + eggs_broken}, Eggs broken {eggs_broken}"
    return result 
```

```python
omelet()
```

```python
print(omelet(3, 4))
```

```python
print(omelet(6))
print(omelet(45))
print(omelet(2))
```

```python
print(omelet(eggs_broken=3, eggs_number=4))
```

```python
print(omelet(4, 3))
```

## `return` vs `print()`

Algo importante que hay que tomar en cuenta es la diferencia entre `return` y `print()`.

* La declaración `return` es usada dentro de una función o método para devolver de vuelta un resultado.
  * En caso de que no haya una declaración `return`, devuelve un `None`.
  * Permite guardar este resultado en una variable y usarla después.

* La función `print()` imprime el resultado para mostrarlo, y se puede utilizarlo en una función en vez de `return`.
  * Sin embargo, no es recomendable utilizarlo.
      * La razón: devuelve un valor `None` extra en ciertos casos y no permite realizar operaciones posteriores.

```python
print(None)
```

```python
type(None)
```

### Ejemplo 19
Veamos la diferencia entre `print()` y `return`.

```python
def f1(x):
  return (x/2 + 1)

def f2(x):
  print(x/2 + 1)
```

Cuando llamamos a la función, devuelven los resultados sin problema.

```python
f1(5)
```

```python
f2(5)
```

Pero, ¿qué pasa cuando se utiliza la función `print()` en caso de que tengamos un código largo?

```python
print(f1(10))
print(f1(2))
```

```python
print(f2(10))
print(f2(2))
```

Devuelve el resultado, ya que dentro de la función está un `print()`, pero también devuelve un `None` porque no hay la declaración `return` en la función.

Y ahora, ¿qué pasa si queremos realizar un cálculo?

```python
y = 4 + f1(6)
print(y)
```

```python
y = 4 + f2(6)
print(y)
```

Python nos indica que no puede hacer una operación con un tipo de dato `int` y valor nulo (`None`).

Por eso es importante solo utilizar `return` para devolver resultados en una función.

### Ejemplo 21
Escribe una función `decompose(n)` que reciba como argumento una cantidad de segundos `n`, y devuelva cuántos días `d`, horas `h`, minutos `m` y segundos `s` representa.

Ejemplo, 3661 segundo son 0 días, 1 hora, 1 min y 1 s.

Se debe devolver de la siguiente forma el resultado:
(0, 1, 1, 1)

```python
def descompose(n):
    #función que pone un número en segundos y devuelve días, horas, min y seg
    s = n                     # s es igual a n, número ingresado
    m = s//60                 # se divide s para 60 (división entera), se fija en m
    s %= 60                   # módulo s de 60 (la diferencia sobrante al momento de dividir)
    h = 0                     # h que es hora se fija en 0
    d = 0                     # d que es día se fija en 0

    # verificación
    if m > 60:                # si m es mayor a 60 ocurre lo de abajo
        h = m//60             # división entera de m, se fija en h
        m %= 60               # módulo m para 60
        if h > 24:            # si h es mayor a 24 va a ocurrir lo de abajo
            d = h//24         # división entera de h se guarda en d
            h %= 60           # módulo de h para 60
    return (d, h, m, s)
```

```python
print(descompose(67))
print(descompose(147))
print(descompose(3661))
print(descompose(76234))
print(descompose(432000))
print(descompose(79083251))
print(descompose(4523465))
```

### Ejemplo 22
Diseñemos una función `delNumbers(str)` que entre un string `str` y devuelva el string pero sin sus caracteres numéricos o dígitos.

```python
print("")
```

```python
# función
def delNumbers(string):
    new_string = ""                   # string vacío
    for letter in string:             # for loop que itera cada letra del argumento ingresado
        if letter.isnumeric():        # método que comprueba si es número o no
            new_string += ""          # se actualiza en el nuevo string y pone algo vacío
        else:
            new_string += letter      # si no, pone las letras normales
    return new_string
```

```python
print(delNumbers("Alicante a 20 kms"))
print(delNumbers("Casa_1234, Bcn25"))
print(delNumbers("Urb. Pedro Jose, OE5 - 364, Código 0DF34gd8d7F"))
```
